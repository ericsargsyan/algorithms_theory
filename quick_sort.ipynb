```python
def partition(array, p, q):
    x = array[p]
    i = p
    
    for j in range(p, q + 1):
        if x > array[j]:
            i = i + 1
            array[i], array[j] = array[j], array[i]
    
    array[p], array[i] = array[i], array[p]
    
    return i
    

def quick_sort(array, p, q):
    
    if len(array) == 1:
        return array
    
    if p < q:
        k = partition(array, p, q)
        quick_sort(array, p, k - 1)
        quick_sort(array, k + 1, q)
        
    return array    
```

```python
array = [6, 11, 15, 5, 4, 9, 3]
quick_sort(array, 0, len(array) - 1)
```
