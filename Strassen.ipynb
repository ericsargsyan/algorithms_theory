```python
import numpy as np
```

```python
def Split(array):
    
    row, col = array.shape
    row1 = row // 2
    col1 = col // 2
    return array[:row1, :col1], array[:row1, col1:], array[row1:, :col1], array[row1:, col1:]
```

```python
def Strassen(A, B):
    
    if len(A) == 1:
        return A * B
    
    A11, A12, A21, A22 = Split(A)
    B11, B12, B21, B22 = Split(B)
    
    M1 = Strassen((A11 + A22), (B11 + B22))
    M2 = Strassen((A21 + A22), B11)
    M3 = Strassen(A11, (B12 - B22))
    M4 = Strassen(A22, (B21 - B11))
    M5 = Strassen((A11 + A12), B22)
    M6 = Strassen((A21 - A11), (B11 + B12))
    M7 = Strassen((A12 - A22), (B21 + B22))
    
    C11 = M1 + M4 - M5 + M7
    C12 = M3 + M5
    C21 = M2 + M4
    C22 = M1 - M2 + M3 + M6
    
    return np.vstack([np.hstack([C11, C12]), np.hstack([C21, C22])])
```

```python
A = np.vstack([np.hstack([1, 2]), np.hstack([-1, 1])])
B = np.vstack([np.hstack([2, 3]), np.hstack([-1, 0])])
```

```python
Strassen(A, B)
```
