```python
def insertion_sort(array):
    if len(array) <= 1:
        return array
    
    for i in range(1, len(array)):
        key = array[i]
        j = i - 1
        while j >= 0 and array[j] > key:
            array[j + 1] = array[j]
            j -= 1
        array[j + 1] = key
        
    return array    
```

```python
a = [9, 8, 7, 6, 5, 4, 3, 2, 1]
b = [1, 2, 3, 4, 6, 5, 7, 8, 9]
insertion_sort(a)
```
