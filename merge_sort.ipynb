```python
def merge_sort(array):
    
    n = len(array)
    
    if n > 1:
        
        mid = n // 2
        A = array[:mid]
        B = array[mid:]

        merge_sort(A)
        merge_sort(B)
    
        i = j = k = 0
        
        while i < len(A) and j < len(B):
            
            if A[i] < B[j]:
                array[k] = A[i]
                i += 1
            else:
                array[k] = B[j]
                j += 1
                
            k += 1    
            
        while i < len(A):
            array[k] = A[i]
            i += 1
            k += 1
                
        while j < len(B):
            array[k] = B[j]
            j += 1
            k += 1
                      
        return array               
```

```python
array = [9, 8, 7, 6, 5, 4, 3, 2, 1]
merge_sort(array)
```
