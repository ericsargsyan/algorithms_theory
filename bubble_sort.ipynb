```python
def buuble_sort(array):
    
    if len(array) <= 1:
        return array
    
    for i in range(len(array) - 1):
        swapped = True
        
        for j in range(len(array) - i - 1):
            if array[j] > array[j + 1]:
                array[j], array[j + 1] = array[j + 1], array[j]
                swapped = False
                
        if swapped:
            return array
```

```python
a = [9, 8, 7, 6, 5, 4, 3, 2, 1]
b = [3, 5, 0, 12, 56, 4, -1, 1]
c = [2, 1, 3, 4, 5, 6, 7, 8, 9]
buuble_sort(c)
```
